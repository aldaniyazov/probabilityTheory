Выбрать тест и проверить, есть ли различия между выборками:

1 )  Даны две  независимые выборки. Не соблюдается условие нормальности
x1  380,420, 290
y1 140,360,200,900

Сделайте вывод по результатам, полученным с помощью функции

```python
from scipy.stats import mannwhitneyu

def mann_whitney_u_test(data1, data2, alpha=0.05):
    # Проведение теста Манна-Уитни
    statistic, p_value = mannwhitneyu(data1, data2)
    
    # Вывод результатов
    print("Статистика теста:", statistic)
    print("p-значение:", p_value)

    # Проверка статистической значимости
    if p_value < alpha:
        print("Есть статистически значимые различия между выборками.")
    else:
        print("Нет статистически значимых различий между выборками.")

# Заданные данные
x1 = [380, 420, 290]
y1 = [140, 360, 200, 900]

# Проверка теста Манна-Уитни на заданных данных
mann_whitney_u_test(x1, y1)
```

2 ) Исследовалось влияние препарата на уровень давления пациентов. Сначала измерялось давление до приема препарата, потом через 10 минут и через 30 минут. Есть ли статистически значимые различия?

1е измерение до приема препарата: 150, 160, 165, 145, 155

2е измерение через 10 минут: 140, 155, 150,  130, 135

3е измерение через 30 минут: 130, 130, 120, 130, 125

```python
from scipy.stats import kruskal

def kruskal_wallis_test(*args, alpha=0.05):
    # Проведение теста Краскела-Уоллиса
    statistic, p_value = kruskal(*args)
    
    # Вывод результатов
    print("Статистика теста:", statistic)
    print("p-значение:", p_value)

    # Проверка статистической значимости
    if p_value < alpha:
        print("Есть статистически значимые различия между группами.")
    else:
        print("Нет статистически значимых различий между группами.")

# Заданные данные
before = [150, 160, 165, 145, 155]
after_10_minutes = [140, 155, 150, 130, 135]
after_30_minutes = [130, 130, 120, 130, 125]

# Проверка теста Краскела-Уоллиса на заданных данных
kruskal_wallis_test(before, after_10_minutes, after_30_minutes)
```

3 ) Сравните 1 и 2 е измерения, предполагая, что 3го измерения через 30 минут не было.

```python
from scipy.stats import wilcoxon

def wilcoxon_signed_rank_test(data1, data2, alpha=0.05):
    # Проведение теста Уилкоксона для парных выборок
    statistic, p_value = wilcoxon(data1, data2)
    
    # Вывод результатов
    print("Статистика теста:", statistic)
    print("p-значение:", p_value)

    # Проверка статистической значимости
    if p_value < alpha:
        print("Есть статистически значимые различия между выборками.")
    else:
        print("Нет статистически значимых различий между выборками.")

# Заданные данные
before = [150, 160, 165, 145, 155]
after_10_minutes = [140, 155, 150, 130, 135]

# Проверка теста Уилкоксона на заданных данных
wilcoxon_signed_rank_test(before, after_10_minutes)
```

4) Даны 3 группы  учеников плавания.
В 1 группе время на дистанцию 50 м составляют: 56, 60, 62, 55, 71, 67, 59, 58, 64, 67

Вторая группа : 57, 58, 69, 48, 72, 70, 68, 71, 50, 53

Третья группа: 57, 67, 49, 48, 47, 55, 66, 51, 54

```python
from scipy.stats import kruskal

def kruskal_wallis_test(*args, alpha=0.05):
    # Проведение теста Краскела-Уоллиса
    statistic, p_value = kruskal(*args)
    
    # Вывод результатов
    print("Статистика теста:", statistic)
    print("p-значение:", p_value)

    # Проверка статистической значимости
    if p_value < alpha:
        print("Есть статистически значимые различия между группами.")
    else:
        print("Нет статистически значимых различий между группами.")

# Заданные данные
group1 = [56, 60, 62, 55, 71, 67, 59, 58, 64, 67]
group2 = [57, 58, 69, 48, 72, 70, 68, 71, 50, 53]
group3 = [57, 67, 49, 48, 47, 55, 66, 51, 54]

# Проверка теста Краскела-Уоллиса на заданных данных
kruskal_wallis_test(group1, group2, group3)
```

5) Заявляется, что партия изготавливается со средним арифметическим 2,5 см. Проверить данную гипотезу, если известно, что размеры изделий подчинены нормальному закону распределения. Объем выборки 10, уровень статистической значимости 5%

2.51, 2.35, 2.74, 2.56, 2.40, 2.36, 2.65, 2.7, 2.67, 2.34

```python
from scipy.stats import ttest_1samp

def one_sample_t_test(data, popmean, alpha=0.05):
    # Проведение одновыборочного t-теста
    statistic, p_value = ttest_1samp(data, popmean)
    
    # Вывод результатов
    print("Статистика теста:", statistic)
    print("p-значение:", p_value)

    # Проверка статистической значимости
    if p_value < alpha:
        print("Есть статистически значимое отклонение от утверждаемого среднего.")
    else:
        print("Нет статистически значимого отклонения от утверждаемого среднего.")

# Заданные данные
sample_data = [2.51, 2.35, 2.74, 2.56, 2.40, 2.36, 2.65, 2.7, 2.67, 2.34]
pop_mean = 2.5
alpha = 0.05

# Проверка одновыборочного t-теста на заданных данных
one_sample_t_test(sample_data, pop_mean, alpha)
```
