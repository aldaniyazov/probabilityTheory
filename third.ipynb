Даны значения зарплат из выборки выпускников: 100, 80, 75, 77, 89, 33, 45, 25, 65, 17, 30, 24, 57, 55, 70, 75, 65, 84, 90, 150. Посчитать (желательно без использования статистических методов наподобие std, var, mean) среднее арифметическое, среднее квадратичное отклонение, смещенную и несмещенную оценки дисперсий для данной выборки.

```python
# Значения выборки
salaries = [100, 80, 75, 77, 89, 33, 45, 25, 65, 17, 30, 24, 57, 55, 70, 75, 65, 84, 90, 150]
n = len(salaries)  # количество значений в выборке

# Среднее арифметическое
mean_salary = sum(salaries) / n

# Дисперсия (смещенная оценка)
variance_biased = sum((x - mean_salary) ** 2 for x in salaries) / n

# Несмещенная оценка дисперсии
variance_unbiased = sum((x - mean_salary) ** 2 for x in salaries) / (n - 1)

# Среднее квадратичное отклонение (стандартное отклонение)
std_deviation = math.sqrt(variance_biased)

print("Среднее арифметическое:", mean_salary)
print("Дисперсия (смещенная оценка):", variance_biased)
print("Дисперсия (несмещенная оценка):", variance_unbiased)
print("Среднее квадратичное отклонение:", std_deviation)
```

В первом ящике находится 8 мячей, из которых 5 - белые. Во втором ящике - 12 мячей, из которых 5 белых. Из первого ящика вытаскивают случайным образом два мяча, из второго - 4. Какова вероятность того, что 3 мяча белые?

```python
# Функция для вычисления количества сочетаний C(n, k)
def combinations(n, k):
    return math.factorial(n) / (math.factorial(k) * math.factorial(n - k))

# Количество белых мячей и общее количество мячей в каждом ящике
white_first = 5
total_first = 8
white_second = 5
total_second = 12

# Вероятность того, что из первого ящика вытащат 1 белый мяч, а из второго - 2 белых мяча
prob_case1 = (combinations(white_first, 1) * combinations(total_first - white_first, 1) / combinations(total_first, 2)) * (combinations(white_second, 2) * combinations(total_second - white_second, 2) / combinations(total_second, 4))

# Вероятность того, что из первого ящика вытащат 2 белых мяча, а из второго - 1 белый мяч
prob_case2 = (combinations(white_first, 2) / combinations(total_first, 2)) * (combinations(white_second, 1) * combinations(total_second - white_second, 3) / combinations(total_second, 4))

# Общая вероятность того, что 3 мяча белые
probability_3_white = prob_case1 + prob_case2

print("Вероятность того, что 3 мяча белые:", probability_3_white)
```

На соревновании по биатлону один из трех спортсменов стреляет и попадает в мишень. Вероятность попадания для первого спортсмена равна 0.9, для второго — 0.8, для третьего — 0.6. Найти вероятность того, что выстрел произведен: a). первым спортсменом б). вторым спортсменом в). третьим спортсменом.

```python
# Вероятности попадания для каждого спортсмена
p_hit = [0.9, 0.8, 0.6]
# Вероятность выбора каждого спортсмена
p_shooter = 1/3

# Вычисление общей вероятности попадания в мишень
p_hit_total = sum(p * p_shooter for p in p_hit)

# Вычисление вероятности попадания для каждого спортсмена при условии попадания в мишень (по формуле Байеса)
p_shooter_hit = [p * p_shooter / p_hit_total for p in p_hit]

# Вывод результатов
print("Вероятность того, что выстрел произведен:")
print("а) первым спортсменом:", p_shooter_hit[0])
print("б) вторым спортсменом:", p_shooter_hit[1])
print("в) третьим спортсменом:", p_shooter_hit[2])
```

В университет на факультеты A и B поступило равное количество студентов, а на факультет C студентов поступило столько же, сколько на A и B вместе. Вероятность того, что студент факультета A сдаст первую сессию, равна 0.8. Для студента факультета B эта вероятность равна 0.7, а для студента факультета C - 0.9. Студент сдал первую сессию. Какова вероятность, что он учится: a). на факультете A б). на факультете B в). на факультете C?

```python
# Вероятности поступления на факультеты A и B
p_A_B = p_B_B = 1 / 3
# Вероятность поступления на факультет C
p_C = 2 * p_A_B
# Вероятности сдачи сессии для каждого факультета
p_pass_A = 0.8
p_pass_B = 0.7
p_pass_C = 0.9

# Вычисление вероятности сдачи сессии для факультетов
p_pass = p_pass_A * p_A_B + p_pass_B * p_B_B + p_pass_C * p_C

# Вычисление вероятностей учебы на каждом факультете при условии сдачи сессии
p_A_given_pass = (p_pass_A * p_A_B) / p_pass
p_B_given_pass = (p_pass_B * p_B_B) / p_pass
p_C_given_pass = (p_pass_C * p_C) / p_pass

# Вывод результатов
print("Вероятность того, что студент учится на факультете A:", p_A_given_pass)
print("Вероятность того, что студент учится на факультете B:", p_B_given_pass)
print("Вероятность того, что студент учится на факультете C:", p_C_given_pass)
```

Устройство состоит из трех деталей. Для первой детали вероятность выйти из строя в первый месяц равна 0.1, для второй - 0.2, для третьей - 0.25. Какова вероятность того, что в первый месяц выйдут из строя: а). все детали б). только две детали в). хотя бы одна деталь г). от одной до двух деталей?

```python
# Вероятности выхода из строя каждой детали
p_A = 0.1
p_B = 0.2
p_C = 0.25

# а) Вероятность того, что все детали выйдут из строя
p_all = p_A * p_B * p_C

# б) Вероятность того, что только две детали выйдут из строя
p_two = p_A * p_B * (1 - p_C) + p_A * (1 - p_B) * p_C + (1 - p_A) * p_B * p_C

# в) Вероятность того, что хотя бы одна деталь выйдет из строя
p_at_least_one = 1 - (1 - p_A) * (1 - p_B) * (1 - p_C)

# г) Вероятность того, что выйдет из строя от одной до двух деталей
p_one_or_two = p_at_least_one - p_all

# Вывод результатов
print("а) Вероятность того, что все детали выйдут из строя:", p_all)
print("б) Вероятность того, что только две детали выйдут из строя:", p_two)
print("в) Вероятность того, что хотя бы одна деталь выйдет из строя:", p_at_least_one)
print("г) Вероятность того, что выйдет из строя от одной до двух деталей:", p_one_or_two)
```
