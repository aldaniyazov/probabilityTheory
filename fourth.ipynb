Случайная непрерывная величина A имеет равномерное распределение на промежутке (200, 800]. Найдите ее среднее значение и дисперсию.

```python
# Заданный промежуток
a = 200
b = 800

# Расчет среднего значения
mean = (a + b) / 2

# Расчет дисперсии
variance = ((b - a) ** 2) / 12

# Вывод результатов
print("Среднее значение:", mean)
print("Дисперсия:", variance)
```

О случайной непрерывной равномерно распределенной величине B известно, что ее дисперсия равна 0.2. Можно ли найти правую границу величины B и ее среднее значение зная, что левая граница равна 0.5? Если да, найдите ее.

```python
import math
```

```python
# Заданные значения
a = 0.5
variance = 0.2

# Находим правую границу (b) из уравнения дисперсии
b = math.sqrt(12 * variance) + a

# Находим среднее значение
mean = (a + b) / 2

# Вывод результатов
print("Правая граница (b):", b)
print("Среднее значение:", mean)
```

Непрерывная случайная величина X распределена нормально и задана плотностью распределения f(x) = (1 / (4 * sqrt(2pi))) * exp((-(x+2)**2) / 32)

Найдите:

а). M(X)

б). D(X)

в). std(X) (среднее квадратичное отклонение)

```python
# Заданные значения
mu = -2
variance = 32

# Расчет стандартного отклонения (std)
std_deviation = math.sqrt(variance)

# Вывод результатов
print("а) Среднее значение (M(X)):", mu)
print("б) Дисперсия (D(X)):", variance)
print("в) Стандартное отклонение (std(X)):", std_deviation)
```

Рост взрослого населения города X имеет нормальное распределение.
Причем, средний рост равен 174 см, а среднее квадратичное отклонение равно 8 см.

Какова вероятность того, что случайным образом выбранный взрослый человек имеет рост:

а). больше 182 см

б). больше 190 см

в). от 166 см до 190 см

г). от 166 см до 182 см

д). от 158 см до 190 см

е). не выше 150 см или не ниже 190 см

ё). не выше 150 см или не ниже 198 см

ж). ниже 166 см.

```python
from scipy.stats import norm

# Заданные параметры
mu = 174  # среднее значение
sigma = 8  # стандартное отклонение

# Стандартизация значений
z_182 = (182 - mu) / sigma
z_190 = (190 - mu) / sigma
z_166 = (166 - mu) / sigma
z_158 = (158 - mu) / sigma
z_150 = (150 - mu) / sigma
z_198 = (198 - mu) / sigma

# Расчет вероятностей
# а)
p_a = 1 - norm.cdf(z_182)
# б)
p_b = 1 - norm.cdf(z_190)
# в)
p_v = norm.cdf((190 - mu) / sigma) - norm.cdf((166 - mu) / sigma)
# г)
p_g = norm.cdf((182 - mu) / sigma) - norm.cdf((166 - mu) / sigma)
# д)
p_d = norm.cdf((190 - mu) / sigma) - norm.cdf((158 - mu) / sigma)
# е)
p_e = norm.cdf((150 - mu) / sigma) + (1 - norm.cdf((190 - mu) / sigma))
# ё)
p_yo = norm.cdf((150 - mu) / sigma) + (1 - norm.cdf((198 - mu) / sigma))
# ж)
p_zh = norm.cdf((166 - mu) / sigma)

# Вывод результатов
print("а) Вероятность того, что рост будет больше 182 см:", p_a)
print("б) Вероятность того, что рост будет больше 190 см:", p_b)
print("в) Вероятность того, что рост будет от 166 см до 190 см:", p_v)
print("г) Вероятность того, что рост будет от 166 см до 182 см:", p_g)
print("д) Вероятность того, что рост будет от 158 см до 190 см:", p_d)
print("е) Вероятность того, что рост будет не выше 150 см или не ниже 190 см:", p_e)
print("ё) Вероятность того, что рост будет не выше 150 см или не ниже 198 см:", p_yo)
print("ж) Вероятность того, что рост будет ниже 166 см:", p_zh)
```

На сколько сигм (средних квадратичных отклонений) отклоняется рост человека, равный 190 см, от математического ожидания роста в популяции, в которой M(X) = 178 см и D(X) = 25 кв.см?

```python
# Заданные параметры
x = 190  # наблюдаемое значение роста
mu = 178  # среднее значение роста
variance = 25  # дисперсия роста

# Стандартное отклонение
std_deviation = math.sqrt(variance)

# Расчет количества сигм
num_sigmas = (x - mu) / std_deviation

# Вывод результата
print("Рост человека отклоняется на", num_sigmas, "средних квадратичных отклонения от среднего значения в популяции.")
```
